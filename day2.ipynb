```python
# prompt: create scatter plot btween CRIM and MEDV column

crim = np.array(data.CRIM)
medv = np.array(data.MEDV)
plt.scatter(crim, medv)
plt.xlabel("CRIM")
plt.ylabel("MEDV")
plt.title("Scatter Plot of CRIM vs MEDV")
plt.show()
```

```python
data.columns
```

```python
# prompt: ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
#        'PTRATIO', 'B', 'LSTAT', 'MEDV']
# create MEDV vs all features scatter plot one by one

features = data.columns

for feature in features:
    # Extract data
    x = np.array(data[feature])
    y = np.array(data['MEDV'])

    # Create scatter plot
    plt.scatter(x, y)
    plt.xlabel(feature)
    plt.ylabel("MEDV")
    plt.title(f"Scatter Plot of {feature} vs MEDV")
    plt.show()
```
